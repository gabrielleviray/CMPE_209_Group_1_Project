```python
from functions import kddDfEncode
from functions import kddReducetoFour
from functions import kddMutualInfo
from functions import saveList
from functions import retainFeatures
from functions import analyze

import pandas as pd
import numpy as np




dataset_file = 'datasets/kddcup.data_10_percent_corrected.csv'
feature_names = pd.read_csv('datasets/feature_names.csv', header=None)
kdd_names = feature_names.iloc[:,0].values

kdd = pd.read_csv(dataset_file, names=kdd_names)
```

```python
kddLabelReduced = kddReducetoFour(kdd)
```

```python
kddLabelReduced
```

```python
X = kddLabelReduced.drop(columns=['attack_type'])
X = pd.get_dummies(X, columns=['protocol_type','flag','service'])
Y = kddLabelReduced['attack_type']
```

```python
X
```

```python
from sklearn import preprocessing
scaler1 = preprocessing.StandardScaler().fit_transform(X.values)
X_scaled = pd.DataFrame(scaler1, index=X.index, columns=X.columns)
X_scaled
```

```python
Y_encoded = pd.get_dummies(Y, columns=['attack_type'])
```

```python
Y_encoded
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import mutual_info_classif


X_train,X_test,y_train,y_test=train_test_split(X_scaled, Y_encoded, test_size=0.75, random_state=42)
```

```python
X_train
```

```python
mutual_info_normal = mutual_info_classif(X_train, y_train['normal.'])
mutual_info_normal = pd.Series(mutual_info_normal)
mutual_info_normal.index = X_train.columns
```

```python
mutual_info_dos = mutual_info_classif(X_train, y_train['dos'])
mutual_info_dos = pd.Series(mutual_info_dos)
mutual_info_dos.index = X_train.columns
```

```python
mutual_info_u2r = mutual_info_classif(X_train, y_train['U2R'])
mutual_info_u2r = pd.Series(mutual_info_u2r)
mutual_info_u2r.index = X_train.columns
```

```python
mutual_info_r2l= mutual_info_classif(X_train, y_train['R2L'])
mutual_info_r2l = pd.Series(mutual_info_r2l)
mutual_info_r2l.index = X_train.columns
```

```python
mutual_info_probe= mutual_info_classif(X_train, y_train['probe'])
mutual_info_probe = pd.Series(mutual_info_u2r)
mutual_info_probe.index = X_train.columns
```

```python
mutual_info_normal.to_csv('data_analytics/MutualInfoNormal.csv', columns=None, header=True, index=True)
mutual_info_dos.to_csv('data_analytics/MutualInfoDos.csv', columns=None, header=True, index=True)
mutual_info_u2r.to_csv('data_analytics/MutualInfoU2R.csv', columns=None, header=True, index=True)
mutual_info_r2l.to_csv('data_analytics/MutualInfoR2L.csv', columns=None, header=True, index=True)
mutual_info_probe.to_csv('data_analytics/MutualInfoProbe.csv', columns=None, header=True, index=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled, Y, test_size=0.75, random_state=42)
mutual_info_all= mutual_info_classif(X_train, y_train)
mutual_info_all = pd.Series(mutual_info_all)
mutual_info_all.index = X_train.columns
mutual_info_all.to_csv('data_analytics/MutualInfoAll.csv', columns=None, header=True, index=True)
```
